```python
from google import genai
client = genai.Client(api_key="Your_api_key")
print("Gemini Connected!")
```

```python
user_query = input("Ask your question: ")
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = user_query
)
print(response.text)
```

```python
# Prompt Template
user_query = input("Ask your question: ")
prompt = f'''
You are an AI mentor helping students prepare for hackathons.
Question : {user_query}
Give :-
1. Explanation
2. Difficulty level
3. Recommended Tech Stack
4. Learning sources
'''
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```

```python
# Role Prompting 
user_query = input("Ask your question: ")
prompt = f'''
You are an expert Hackathon mentor, your job is to help students build practical AI projects.
Question : {user_query}
Answer strictly in the following format:
1. Project Title:
2. Problem Statement: 
3. Key Features:
4. Tech Stack:
5. Difficulty (easy/medium/hard):
6. Learning Outcome:
7. Hackathon Score (out of 10):
Do not skip any section
'''
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```
