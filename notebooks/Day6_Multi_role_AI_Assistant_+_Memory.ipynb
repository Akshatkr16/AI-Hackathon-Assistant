```python
from google import genai
client = genai.Client(api_key = "Your_api_key")
print("Gemini Connected!")
```

```python
# Storing chat history
chat_history = []
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("Goodbye !")
        break
    # store user message
    chat_history.append(
        f"User: {user_query}"
    )
    response = client.models.generate_content(
        model = "gemini-2.5-flash",
        contents = "\n".join(chat_history) # send full history
    )
    # store assistant resposne
    chat_history.append(
        f"Assistant: {response.text}"
    )
    print("\nAssistant: ")
    print(response.text)
    print("-"*150)
```

```python
# Understanding multiple modes of role and role prompt
user_query = input("Ask your question: ")
prompt = f'''
You are a CEO.
Focus on business impact.
Question : {user_query}
'''
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```

```python
# create different modes
mode = input('''
Choose Assistant mode:
       1. Supply Chain Assistant
       2. Data Analyst
       3. Mechanical Engineering
    Enter Choice (1 or 2 or 3): 
''')
# create role prompt
if mode == 1:
    role_prompt = '''
                  You are a senior Supply Chain consultant
                  Help users with logistics, inventory, procurement and forecasting
                  Give:
                      1. Explanation
                      2. Recommendation
                      3. Business Impact
    '''
elif mode == 2:
    role_prompt = '''
                  You are a senior Data Analyst
                  Help users with SQL, Excel, PowerBI, Python and Analytics
                  Give:
                      1. Explanation
                      2. Example
                      3. Best Practices
    '''
elif mode == 3:
    role_prompt = '''
                  You are a senior Mechanical Engineering
                  Help users with manufacturing, thermodynamics, fluids mechanics and machinery, machine design and production
                  Give:
                      1. Explanation
                      2. Engineering Perspective
                      3. Practical Applications
    '''
else:
    role_prompt = 'You are a helpful AI Assistant !'
# initialize chat history
chat_history = [role_prompt]
# start memory + chat loop
while True:
    user_query = input("\nYou: ")
    if user_query.lower() == 'exit':
        print("Good Bye!")
        break
    # store user message
    chat_history.append(f"User: {user_query}")
    try:
        response = client.models.generate_content(
            model = "gemini-2.5-flash",
            contents = "\n".join(chat_history)
        )
        # store assistant response
        chat_history.append(f"Assiatant: {response.text}")
        print("\nAssistant:\n")
        print(response.text)
        print("-"*150)
    except Exception as e:
        print("\nError Occured:")
        print(e)
```
