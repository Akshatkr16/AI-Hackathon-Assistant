```python
from google import genai
client = genai.Client(api_key = "Your api_key")
print("Gemini Connected!")
```

```python
# Creating Chat Loop
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("Goodbye !")
        break
    try:
        response = client.models.generate_content(
        model = "gemini-2.5-flash",
        contents = user_query
        )
        print("\nAssistant:")
        print(response.text)
        print("-"*50)
    except Exception as e:
        print("Error: ",e)
```

```python
# Assistant Persona
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("Good Bye!")
        break
    prompt = f'''
    You are a senior Supply Chain Consultant.
    Help the user solve supply chain, inventory, logistics, procurement & forecasting problems 
    User Question: {user_query}
    Give:
    1. Explanation
    2. Recommendation
    3. Buisness Impact
    '''
    response = client.models.generate_content(
        model = "gemini-2.5-flash",
        contents = prompt
    )
    print("\n SupplyChain AI: ")
    print(response.text)
    print("-"*100)
```
