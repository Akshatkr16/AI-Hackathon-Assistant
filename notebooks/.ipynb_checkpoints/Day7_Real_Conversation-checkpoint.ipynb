```python
from google import genai
client = genai.Client(api_key = "Your_api_key")
print("Gemini Connected!")
```

```python
chat_history = ""
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("Good bye!")
        break
    # Build full prompt
    full_prompt = f'''
                   You are a Senior Supply Chain consultant.
                   Give Professional supply chain advice
                   Chat_History: {chat_history}
                   User: {user_query}
    '''
    response = client.models.generate_content(
        model = "gemini-2.5-flash",
        contents = full_prompt                    # send full prompt
    )
    print("\nAssistant:")
    print(response.text)
    # update memory
    chat_history += f"\nUser: {user_query}"
    chat_history += f"\nAssistant: {response.text}\n"
    print("-"*150)
```
