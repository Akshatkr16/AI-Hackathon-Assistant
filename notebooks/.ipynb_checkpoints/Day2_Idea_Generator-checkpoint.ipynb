```python
from google import genai
client = genai.Client(api_key ="Your_API_Key")
domain = "Supply Chain"
prompt = f'''
Generate ONE hackathon project idea in the domain: {domain}

Format your response exactly as:

PROJECT TITLE:
...

PROBLEM STATEMENT:
...

KEY FEATURES:
- ...
- ...
- ...

TECH STACK:
- Frontend:
- Backend:
- Database:
- AI Model:

DIFFICULTY:
Easy/Medium/Hard
'''
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```
