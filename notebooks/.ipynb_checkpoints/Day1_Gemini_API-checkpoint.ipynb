```python
from google import genai
client = genai.Client(
    api_key="Your_API_Key"
)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Give me 3 AI hackathon project ideas"
)

print(response.text)
```

```python
# Trying Better prompts
prompt = '''
          Suggest 5 supply chain AI Projects
          For each project provide:
          1. Problem
          2. Features
          3. Tech Stack
          4. Difficulty
'''
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```
