```python
domain = input("Enter domain: ")
team_size = input("Enter team size: ")
duration = input("Enter hackathon duration (hours): ")
budget = input("Enter budget (low/medium/high): ")
```

```python
prompt = f"""
You are an expert hackathon mentor.
Generate one hackathon project idea.
Constraints:
Domain: {domain}
Team Size: {team_size}
Duration: {duration} hours
Budget: {budget}
Return exactly in this format:
PROJECT TITLE:
PROBLEM STATEMENT:
WHY THIS PROJECT IS SUITABLE:
FEATURES:
- Feature 1
- Feature 2
- Feature 3
TECH STACK:
- Frontend
- Backend
- Database
- AI Model
IMPLEMENTATION PLAN:
Day 1:
Day 2:
Day 3:
DIFFICULTY:
"""
```

```python
from google import genai
client = genai.Client(api_key = "Your_API_Key")
response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt
)
print(response.text)
```

```python
# ask follow up
followup = input("Ask a followup question: ")
new_prompt = f'''
Project Idea: {response.text},
User question: {followup}
Answer the question in detail
'''
```

```python
follow_response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = new_prompt
)
print(follow_response.text)
```
